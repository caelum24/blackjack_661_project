```python
import torch
from split_train_agent import train_agent
from modeling import model
from DQNAgent import DQNAgent
from split_environment import BlackjackEnv
from Supervised_betting import collect_dictionary
from Supervised_betting import evaluate_nn_betting_agent
from BettingModel import BettingNN
from BettingModel import train_betting_nn
import pickle
from betting_grid_search import betting_grid_search

count_type = "full"
agent = DQNAgent("full")
env = BlackjackEnv(count_type = "full")

checkpoint = torch.load("playing_model.pth", weights_only = False)
agent.policy_net.load_state_dict(checkpoint['policy_net_state_dict'])
agent.target_net.load_state_dict(checkpoint['target_net_state_dict'])
agent.optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
agent.epsilon = checkpoint['epsilon']

dictionary = collect_dictionary(agent, env, count_type, 1000000)
#with open("count_reward_dict.pkl", "wb") as f:
    #pickle.dump(dictionary, f)

#with open("count_reward_dict.pkl", "rb") as f:
    #dictionary = pickle.load(f)

model = train_betting_nn(dictionary, count_type, batch_size=64, epochs=10)
evaluate_nn_betting_agent(model, agent, env, count_type, episodes = 10000, max_bet = 1000, min_bet = 1)

print("Betting agent trained and saved.")
```
